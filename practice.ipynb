```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
%matplotlib inline
```

```python
teens = pd.read_csv('C:/Users/USER/Dropbox/kaggle/snsdata.csv')
print(teens.head())
```

```python
#利用lambda 表達式設立dataframe搜尋條件
teens.loc[lambda df : df['friends'] > 500, :]
```

```python
#自定義函數
def query_data(df):
    return (df['friends'] > 500) & df['gender'].str.startswith('F')
teens.loc[query_data, :]
```

```python
# dateframe apply方法做 series的運算
def personality(x):
    if x['friends'] > 500:
        return '太活潑'
    if x['friends'] > 400:
        return '活潑'
    return '普通'
teens.loc[:, 'personality'] = teens.apply(personality, axis=1)
print(teens.head())
```

```python
# dataframe assign方法做 series的運算
teens = teens.assign(
    gradyear_china = lambda x : x['gradyear'] - 1911
)
teens.head()
```

```python
# dataframe 利用條件分組方法做 series的運算
teens.loc[ teens['age'] >= 18, 'type' ] = '成年'
teens.loc[ teens['age'] < 18, 'type' ] = '未成年'
teens.head(5)
```
