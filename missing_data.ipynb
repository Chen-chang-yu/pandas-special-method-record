```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel(r'C:\Users\USER\Dropbox\kaggle\score.xlsx', engine='openpyxl', skiprows = 1)
df.head()
```

```python
# all的用法是代表全部符合空值才進行刪除 
df.dropna(axis = 1, how ='all', inplace =True)
df
```

```python
# all的用法是代表全部符合空值才進行刪除 
df.dropna(axis = 0, how ='all', inplace =True)
df
```

```python
# 將 missing的分數填上 0, 但這不會取代原先的 df, 所以還要額外加工
df['分數'].fillna(0)
```

```python
df['分數'] = df['分數'].fillna(0)
df
# 快速解, 可得到一樣的效果
#df.fillna({'分數' : 0})
```

```python
# 將姓名遺缺值填滿, 利用 forward fill(代碼 : ffill)的方法
df['姓名'] = df['姓名'].fillna(method = 'ffill')
df
```
